```python
# 1. Compute the average ClaimAmount per Province.
ans1 = pd.pivot_table(df , index= 'Province',values='ClaimAmount',aggfunc ='mean')
ans1

# 2.Compute the total Premium per Province.
ans2 = pd.pivot_table(df , index= 'Province' , values= 'Premium' , aggfunc= 'sum')
ans2

# 3. Count the number of policies per Province.
ans3 = pd.pivot_table(df, values= 'PolicyType', index= 'Province' , aggfunc= 'count')
ans3 

# 4. Compute the average ClaimAmount per VehicleType.
ans4 = pd.pivot_table(df , values= 'ClaimAmount' , index= 'VehicleType' , aggfunc= 'mean' )
ans4 

# 5. Find the maximum Premium per Gender.
ans5 = pd.pivot_table(df , values= 'Premium' , index = 'Gender' , aggfunc= 'max' )
ans5 
```

```python
#  1. Compute the average ClaimAmount per Province and VehicleType.

ans1 = pd.pivot_table(df , index= ['Province' , 'VehicleType'] , values = 'ClaimAmount' , aggfunc= 'mean' )
ans1 

# 2. Compute the total Premium per Province and VehicleType.
ans2 = pd.pivot_table(df , index= ['Province' , 'VehicleType'] , values = 'Premium' , aggfunc= 'sum' )
ans2

# 3. Compute the average ClaimAmount per Province split by Gender.
ans3 = pd.pivot_table(df , index= ['Province'] , columns= 'Gender' , values = 'ClaimAmount' , aggfunc= 'mean' )
ans3

# 4. Compute the sum of Premium per Province split by Gender.
ans4 = pd.pivot_table(df , index= ['Province'] , columns= 'Gender' , values = 'Premium' , aggfunc= 'sum' )
ans4

# 5. Count the number of policies per Province and PolicyType.
ans5 = pd.pivot_table(df , index= ['Province', 'PolicyType']   , aggfunc= 'size' )
ans5
```

```python
# 1. Compute the average ClaimAmount and Premium per Province & VehicleType, split by Gender.
ans1 = pd.pivot_table(df , values= ['ClaimAmount' ,'Premium'] , index= ['Province' ,'VehicleType' ] ,columns= 'Gender' , aggfunc= 'mean' )
ans1

#2. Compute ClaimAmount (mean) and Premium (sum) per Province & VehicleType, split by Gender.
ans2 = pd.pivot_table(df , values= ['ClaimAmount' ,'Premium'] , index= ['Province' ,'VehicleType' ] ,columns= 'Gender' , aggfunc= {'ClaimAmount':'mean' , 'Premium':'sum'} )
ans2

# 3. Include totals (margins=True) for the above pivot table.
ans3 = pd.pivot_table(df , values= ['ClaimAmount' ,'Premium'] , index= ['Province' ,'VehicleType' ] ,columns= 'Gender' , margins= True, aggfunc= {'ClaimAmount':'mean' , 'Premium':'sum'} )
ans3

# 4. Replace missing values with 0 in the pivot table above.
ans4 = pd.pivot_table(df , values= ['ClaimAmount' ,'Premium'] , index= ['Province' ,'VehicleType' ] ,fill_value= 0,columns= 'Gender' , margins= True, aggfunc= {'ClaimAmount':'mean' , 'Premium':'sum'} )
ans4

# 5. 
'''
Create a pivot table with multi-level aggregation:
ClaimAmount → mean, max
Premium → sum, min
Rows → Province + VehicleType
Columns → Gender'''
ans5 = pd.pivot_table(df , index= ['Province' ,'VehicleType'] , values= ['ClaimAmount' ,'Premium' ] ,fill_value= 0,columns= 'Gender' , aggfunc= {'ClaimAmount':['mean' , 'max'], 'Premium':['sum', 'min']} )
ans5
```

```python
# 1. Compute the average ClaimAmount per Year and Province, split by Gender, include totals.
ans1 = pd.pivot_table(df , index= ['Year','Province'],columns= 'Gender',margins= True,values='ClaimAmount',aggfunc='mean' )
ans1

# 2. Compute the total Premium per Year, Province, VehicleType, and split by PolicyType.
ans2 = pd.pivot_table(df , values= 'Premium' ,index= ['Year', 'Province', 'VehicleType'] , aggfunc= 'sum', columns= 'Gender')
ans2

# 3. Identify Province and VehicleType combinations with no policies for a Gender, using fill_value=0.
ans3 = pd.pivot_table(df , index= ['Province' , 'VehicleType'] , columns= 'Gender' ,fill_value=0 , values= 'PolicyType' , aggfunc= 'count' )
no_pol = ans3[(ans3==0).any(axis=1)]
no_pol
```
