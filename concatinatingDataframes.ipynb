```python
# Concatenating  DataFrame with pandas
# concatenating DataFrames is a common operation in data analysis, and
# pandas provides a powerful function called `concat()` to achieve this. 
# The `concat()` function can be used to concatenate DataFrames along a 
# particular axis (either rows or columns).

import pandas as pd 

df1 = pd.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]} , index = ['a', 'b', 'c'])
df2 = pd.DataFrame({"D": [7,8,9], "E": [10,11,12]} , index = ['d', 'e', 'f'])
print(df1)
print(df2)

# what parameters does the concat function take?
# The `concat()` function takes several parameters, including:
# - `objs`: The list of DataFrames to concatenate
# - `axis`: The axis along which to concatenate (0 for rows, 1 for columns)
# - `join`: How to handle indexes (default is 'outer')
# - `ignore_index`: Whether to ignore the original index (default is False)
# - `sort`: Whether to sort the non-concatenation axis (default is False)
# pd.concat(objs, axis=0 , join='outer', ignore_index=False)

# Concatenating along rows (axis=0) or one below the other 
df_concat_rows = pd.concat([df1, df2], axis=0, join = 'outer', ignore_index=False )
print(df_concat_rows)

# Concatenating along columns (axis=1) or side by side 
df_concat_cols = pd.concat([df1, df2], axis=1, join = 'outer', ignore_index=False )
print(df_concat_cols)

# with keys parameter
# The `keys` parameter allows you to create a hierarchical index for the concatenated DataFrame.
df_concat_keys = pd.concat([df1, df2], axis=0, join = 'outer', ignore_index=False , keys=['df1', 'df2'])
print(df_concat_keys)

df_concat_keys_cols = pd.concat([df1, df2], axis=1, join = 'outer', ignore_index=False , keys=['df1', 'df2'])
print(df_concat_keys_cols)
```

```python
#lets look at concat in details with the parameters
import pandas as pd 
table1  = pd.DataFrame({'ID': [1, 2, 3, 4, 5],
    'Name': ['A', 'B', 'C', 'D', 'E'],
    'Age': [20, 21, 22, 23, 24],
    'City': ['AA', 'BB', 'CC', 'DD', 'EE'],
    'Score': [85, 90, 88, 92, 87],
    'beutiful' : ['yes', 'no', 'yes', 'no', 'yes']}, index = ['a', 'b', 'c', 'd', 'e'])

table2 = pd.DataFrame({'ID': [6, 7, 8, 9, 10],
    'Name': ['F', 'G', 'H', 'I', 'J'],
    'Age': [25, 26, 27, 28, 29],
    'City': ['FF', 'GG', 'HH', 'II', 'JJ'],
    'Score': [80, 86, 89, 91, 84],
    'fire' : [False , True, False, True, False]} , index = [8,9,10,11,12])

print(table1)
print("\n")
print(table2)
```

```python
# check every parameter of the concat function
df = pd.concat([table1 , table2 ] , axis = 0 )
print(df)
```

```python
df = pd.concat([table1 , table2] , axis= 1 )
print(df)
```

```python
df = pd.concat([table1 , table2 ], axis=0, ignore_index=True  )
print(df)
```

```python
df = pd.concat([table1 , table2 ], axis=0, join = 'inner', ignore_index=False )
print(df)
```

```python
df = pd.concat([table1 , table2 ], axis=0, join = 'outer', ignore_index=False , keys= ["table1", "table2"] )
print(df)
```

```python
df = pd.concat([table1 , table2 ], axis=0, join = 'outer', ignore_index=False , keys= ["table1", "table2"], sort=True )
print(df)
```
