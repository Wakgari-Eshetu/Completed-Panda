```python
# Viewing the data set we have
import pandas as pd
import numpy as np

df = pd.DataFrame(
    np.random.randint(1, 100, size=(20, 10)),
    columns=["Age","Score","Height","Weight","Math","Physics","Chem","Bio","Eng","CS"]
)

print(df) 
```

```python
'''The functions we use to view the dataset '''

# head :show the first given row by default 5  
df.head() # or df.head(10)
# tail : shows the last given row by default 5 
df.tail() # or df.tail(9) last 9 rows 
# sample(): return random rows by default 5 
df.sample()
# shape: shows at the no of row and columns 
df.shape
# columns : the name of columns we have 
df.columns
# index : shows as the rows label
df.index 
# dtypes: type of each columns 
df.dtypes
# info() : this can be very help full for columns name , non-null values, data types and memory usage
df.info() 
#describe(): works for numeric values and calculates count , mean , standard devation , minimum , quartiles , maximum
df.describe()
# .unique(): return all different values in a columns  
df["Age"].unique()
# .nunique() : return number of unique values in each columns
df.nunique()
# value_counts(): count the occurrance of each value in a given columns
df["Age"].value_counts()
# values : return the given dataframe to numpy array just the raw data by removing the index and the columns 
print(df.values) # or the same functionality using the "df.to_numpy()"
```
