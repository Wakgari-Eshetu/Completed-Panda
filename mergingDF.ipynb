```python
import pandas as pd

table1 = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [25, 30, 35],
    'city': ['New York', 'Los Angeles', 'Chicago'],
    'country': ['ethiopia', 'nigeria', 'japan']

})

table2 = pd.DataFrame({
    'id': [4, 5, 6],
    'name': ['David', 'Eve', 'Frank'],
    'age': [28, 32, 29],
    'city': ['Boston', 'Seattle', 'Miami'],
    'country': ['canada', 'congo', 'mali']
})

print(table1 , "\n\n\n" , table2)
```

```python
# LETS MERGE THE TWO TABLES TOGETHER USING DIFFERENT METHODS parameters 
df = pd.merge(table1, table2, how='outer', on='id')

# what is pd.merge() ? 
# is a function in the pandas library that allows
# you to combine two DataFrames based on a common column or index.
# It is similar to SQL JOIN operations and provides various options for merging data.

# what are the parameters in  pd.merge() ? 
# parameters in pd.merge() include: 
# pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False)
# what are the different outputs of pd.merge() ?
```

```python
df1 = pd.DataFrame({
    'id': [1, 2, 3],'name': ['Alice', 'Bob', 'Charlie'],
    'age': [25, 30, 35],})
df2 = pd.DataFrame({
    'id': [4, 2, 3],
    'gender':['male', 'female' , 'male'],
    "country": ['ethiopia', 'nigeria', 'japan']})

df = pd.merge( df1, df2 , on = 'id' , how = 'inner')
print(df)

# what will happen if we doesnt specify the on parameter ? the ans is that it will merge on the common columns which is 'id' in this case. 
# which means it choose by it self the common column to merge on. right ? yes it is correct.
# what if we have no common columns ? then it will merge on the index.
# how ? if we have no common columns and we dont specify the on parameter it will merge on the index by default.
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'value': [10, 20, 30]
})

df2 = pd.DataFrame({
    'score': [100, 200, 300]
}, index=[1, 2, 4])  

pd.merge(df1, df2, left_on='id', right_index=True)
```
