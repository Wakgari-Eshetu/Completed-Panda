```python
# working .join method 
import pandas as pd
help(pd.DataFrame.join)
```

```python
table1= pd.DataFrame({
    'id':[1,2,3,4],
    'name':['leli','kena','nige','abishu'],
    'gender':['M', 'M' ,'F', 'M'],
    'grade':[100,200,300,400]

})


table2= pd.DataFrame({
    'id':[1,2,3,5],
    'name':['eli','ena','ige','abi'],
    'gender':['F', 'F' ,'M', 'M'],
    'grade':[10,20,30,40]
} , index= [1,5,2,4])

result = table1.join(table2  ,lsuffix= 'table1' , rsuffix= 'table2' )
print(result)
```
