```python
from  rapidfuzz import fuzz

'''what does rapid fuzz do ?
it is a library that is used to compare two strings and return a score 
between 0 and 100. The higher the score, the more similar the two strings
are. It is used to find duplicates in a dataset.

what does it uses ?
1. Damerau-Levenshtein distance: it is a measure of the difference between two strings. It is defined as the minimum number of single-character edits (insertions, deletions, or substitutions) required to change one string into the other.
2. Jaro-Winkler distance: it is a measure of the similarity between two strings
3. token based similarity: it is a measure of the similarity between two strings based on the tokens (words) in the strings. It is defined as the number of common tokens divided by the total number of tokens in both strings.
'''
```

```python
data = [
    "Abel Tesfaye",
    "Abel Teshfaye",
    "Abel Tesfaie",
    "Abel Tesfay",
    "Amanuel Gebremichael",
    "Amaniel Gebremikael",
    "Amanuel Gebre Michael",
    "Mekdes Alemu",
    "Mekdes Alemmu",
    "Mekdes A.",
    "Addis Ababa University",
    "Addis Abeba University",
    "Adiss Ababa Univ",
    "AAU",
    "Bahir Dar University",
    "Bahir Dar Univ.",
    "BahirDar University"
]

df =pd.DataFrame(data , columns=['Name'], index= range(1,len(data)+1))
df.head()
```

```python
df.duplicated().sum()  # there is no exact duplicatates.
```

```python
from rapidfuzz import utils
cleaned_data = [utils.default_process(name) for name in data]
for first , second in zip(data , cleaned_data):
    print(f"Original: {first} --> Cleaned: {second}")
```
