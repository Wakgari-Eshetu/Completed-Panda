2. Fuzzy Duplicate Detection
    * duplicate but not exactly but refer to same entity

Why it happens:
    1. due to typos: typing error 
    2. due to format: Abisinia and abisinia
    3. due to missing values 
    4. Abbreviations: AA and AddisAbaba
    5. short forms : leli and lelisa 


Techniques:
    1. string similarity 
        A. Levenshtein Distance: how many steps do we need to make string A same as string B [insert , delete , replace ]
        B. 






```python
# when fuzzy duplicates steps to follow 
#1. Text normalization (lowercase, remove punctuation, etc.)
''' lowercasing : changing all text to lowercase to ensure that "John" and "john" are treated as the same.
    Removing spaces : eliminating leading and trailing spaces to avoid discrepancies caused by extra spaces.
    Removing punctuation : stripping out punctuation marks to prevent them from affecting the comparison.
    Standardizing formats : ensuring that dates, phone numbers, and other formatted data are consistent across entries.
    Removing stop words : eliminating common words that do not add significant meaning (e.g., "the", "and") to focus on the core content.
    Stemming or lemmatization : reducing words to their base or root form to treat different'''

# to lowercase 
df['name'] = df['name'].str.lower()
# remove spaces
df['name'] = df['name'].str.strip()
# remove punctuation
df['name'] = df['name'].str.replace(r'[^\w\s]', '', regex= True)
# \w character class that matches any word character (equivalent to [a-z,A-Z,0-9,_])
# \s character class that matches any whitespace character (equivalent to [ \t\n\r\f\v]) space , tab , newline , carriage return , form feed , vertical tab
# standardize formats
df['date'] = pd.to_datetime(df['date'], errors='coerce') # same format for all dates
df['phone'] = df['phone'].str.replace(r'\D', '', regex=True) # \D matches any non-digit character
# remove stop words
stop_words = set(['the', 'and', 'is', 'in', 'to', 'of']) # example stop words
df['name'] = df['name'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
# stemming or lemmatization
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()
df['feedback'] = df['feedback'].apply(lambda x: ' '.join([lemma.lemmatize(word , pos='v') for word in x.split()]))  # example of lemmatization
# when not to use it :for name and address and short forms

  
```

```python
# the more accurate way is to use with the pos tag and word_tokenize

from nltk import WordNetLemmatizer
from nltk import pos_tag , word_tokenize

lemma = WordNetLemmatizer()
def pos_return(exmaple):
    if exmaple.startswith('N'):
        return 'n'
    elif exmaple.startswith('V'):
        return 'v'
    elif exmaple.startswith('J'):
        return 'a'
    elif exmaple.startswith('R'):
        return 'r'
    else:
        return 'n'

def lammatize_text(text):
    tokens = word_tokenize(text)
    pos_tag = pos_tag(tokens)

    return ' '.join([lemma.lemmatize(word , pos= pos_return(pos)) for word , pos in pos_tag])
    
df['feedback'] = df['feedback'].apply(lammatize_text)

```
