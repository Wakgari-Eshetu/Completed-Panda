before checking the missing value we should detect what is missing 

in order to now what is missing fisrt we should check the validity and dataenforcement 

```python
import pandas as pd
import numpy as np  
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'age': [25, np.nan, 30, None],
    'salary': [5000, 6000, '?', 7000],
    'gender': ['F', 'M', np.nan, 'M']
})
df
```

```python
df.info()
```

```python
df['age'] = df['age'].round().astype('Int64')
df['age'].dtypes
print(df['age'])
```

```python
df['salary'] = pd.to_numeric(df['salary'], errors= 'coerce')
print(df['salary'])
```

```python
# detection now takes place
df.isnull()
```

```python
# count missing value per columns 
missing_value_detection = df.isnull().sum()
print(missing_value_detection)
```

```python
# count missing values per row 
df.isnull().sum(axis=1)
```

```python
# we can also detect row which has missing value
# 1. any missing value in row 
# 2. all missing value in row 

df[df.isnull().any(axis=1)]
df[df.isnull().all(axis=1)]
```

```python
# we can also detect columns which has missing value
# 1. any missing value in row 
# 2. all missing value in row 

df[df.isnull().any()]
df[df.isnull().all()]
```
