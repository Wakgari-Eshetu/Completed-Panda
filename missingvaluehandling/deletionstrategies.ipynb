Deleting strategies 
1.litwise: delete every row that contains the missing value[good to use when value is <5%]
2.Pairwise Deletion
Instead of deleting entire rows, we only remove missing values when computing statistics
3.Column Deletion
Remove entire columns that contain too many missing values.[good to use when the column has 40% - 60% missing value ]
4.Threshold-Based Deletion
Remove rows/columns if missing values exceed a threshold.

```python
df = pd.DataFrame({
    'name':['m','n','b','v'],
    'age':[12,23,None,45],
    'grade':[3.4,None ,4,2.3]
} , index= [x for x in range(1,5)])

df
```
