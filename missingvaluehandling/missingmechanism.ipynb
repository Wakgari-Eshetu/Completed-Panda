```python
import pandas as pd
import numpy as np

np.random.seed(0)

data = pd.DataFrame({
    "age": np.random.randint(18, 60, 30),
    "salary": np.random.randint(3000, 10000, 30),
    "department": np.random.choice(["IT", "HR", "Sales", "Finance"], 30),
    "experience": np.random.randint(1, 20, 30),
    "bonus": np.random.randint(500, 5000, 30)
})

# Introduce missing values
data.loc[[2,7,12], "salary"] = np.nan
data.loc[[5,11,20], "department"] = np.nan
data.loc[[3,8,25], "bonus"] = np.nan
data.loc[[10,15], "experience"] = np.nan

print(data)
```

```python
df = data
df.isnull().sum()
```
