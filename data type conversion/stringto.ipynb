```python
import pandas as pd 
help(pd.to_datetime)
```

```python
df = pd.DataFrame({
    'name':['leli', 'kena' , 'nige' , 'eshetu'],
    'dateofbirth':['10/05/2000', '21/03/1999', '15/07/2001', '30/12/1998']
})

df.dtypes 

df['dateofbirth'] = pd.to_datetime(df['dateofbirth'] , errors= 'coerce', format= '%d/%m/%Y')

df.dtypes
print(df)
```

```python
# string to numeric(int/ float)
import pandas  as pd 
df = pd.DataFrame({'name':['leli' , 'kena' , 'whatt','james'],
                   'price':['100', '-200', '300$', '400*']})

df
```

```python
help(pd.to_numeric)
```

```python
df['cleaned'] = df['price'].str.replace(r'[^0-9.]','', regex= True )
df['price'] = pd.to_numeric(df['cleaned'] , errors='coerce')
df
```
