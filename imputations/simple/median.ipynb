```python
import pandas as pd
import numpy as np

# Create a sample dataset
data = {
    'Age': [25, np.nan, 30, 22, np.nan, 28],
    'Salary': [5000, 6000, np.nan, 4500, 7000, np.nan],
    'Gender': ['Male', 'Female', np.nan, 'Male', 'Female', np.nan],
    'City': ['NY', np.nan, 'LA', 'NY', np.nan, 'Chicago']
}

df = pd.DataFrame(data)
print("Original Dataset:")
print(df)
```

```python
# Filling with median 
num_cols = df.select_dtypes(include=['Int64', 'float64'])
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

df
```
