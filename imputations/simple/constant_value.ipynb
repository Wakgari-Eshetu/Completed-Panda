```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'City': ['NY', np.nan, 'LA', 'Chicago', np.nan],
    'NumClaims': [2, np.nan, 0, 1, np.nan]
})



print(df)
```

```python
df.info()
```

```python
# For numerical cols

num_cols = df.select_dtypes(include=['float64'])
for col in num_cols:
    df[col] = df[col].fillna(0)

cat_cols = df.select_dtypes(include=['object'])
for col in cat_cols:
    df[col] = df[col].fillna('unknown')

df
```
