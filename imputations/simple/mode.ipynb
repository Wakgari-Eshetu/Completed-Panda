```python
import pandas as pd
import numpy as np

# Create a sample dataset
data = {
    'Age': [25, np.nan, 30, 22, np.nan, 28],
    'Salary': [5000, 6000, np.nan, 4500, 7000, np.nan],
    'Gender': ['Male', 'Female', np.nan, 'Male', 'Female', np.nan],
    'City': ['NY', np.nan, 'LA', 'NY', np.nan, 'Chicago']
}

df = pd.DataFrame(data)
print("Original Dataset:")
print(df)
```

```python
df['Gender'] = df['Gender'].astype('category')
```

```python
# fill the categorical cols with the most frequent values

cat_cols = df.select_dtypes(include=['object','category'])
for col in cat_cols:
    mode_cols = df[col].mode() # when the elements of the columns has eqaul number 
    df[col] = df[col].fillna(mode_cols[0])

df

```

```python
# when the frequency is equal we can use random again 

import random

cat_cols = df.select_dtypes(include= ['object','category'])
for col in cat_cols:
    mode_cols = df[col].mode()
    df[col] = df[col].fillna(random.choice(mode_cols))

df
```
