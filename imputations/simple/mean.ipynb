```python
import pandas as pd
import numpy as np

# Create a sample dataset
data = {
    'Age': [25, np.nan, 30, 22, np.nan, 28],
    'Salary': [5000, 6000, np.nan, 4500, 7000, np.nan],
    'Gender': ['Male', 'Female', np.nan, 'Male', 'Female', np.nan],
    'City': ['NY', np.nan, 'LA', 'NY', np.nan, 'Chicago']
}

df = pd.DataFrame(data)
print("Original Dataset:")
print(df)
```

```python
# filling null values with the mean 
df['Age'] = df['Age'].astype('Int64')
print(df.dtypes)
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(int(df['Age'].mean()))
df
```

```python
help(df.select_dtypes)
```

```python
# when to use mean is when the data type is numerical
df.info()
num_col = df.select_dtypes(include=['Int64' ,'float64'])
print(num_col)
 
# Now imputations with the mean 

for col in num_col:
    df[col] = df[col].fillna(df[col].mean())

df
```
