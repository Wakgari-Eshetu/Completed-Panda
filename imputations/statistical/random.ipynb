```python
import pandas as pd
import numpy as np

np.random.seed(42)  # for reproducibility

# 20 rows, 5 columns
n_rows = 20
n_cols = 5

# Generate numeric data
data = {
    f"feature{i+1}": np.random.randint(10, 100, n_rows).astype(float) 
    for i in range(n_cols)
}

df = pd.DataFrame(data)

# Introduce ~20% random missing values in each column
for col in df.columns:
    missing_indices = np.random.choice(df.index, size=4, replace=False)  # 4 ~ 20%
    df.loc[missing_indices, col] = np.nan

print("Dataset with missing values:\n")
print(df)
```

```python
lists = df.columns[df.isnull().any()].tolist()
for col in lists:
    non_null_values = df[col].dropna()
    missing_values  = df[col].isnull().sum()
    random_sample = non_null_values.sample(n= missing_values , replace= True ,random_state =  42 )
    df.loc[df[col].isnull(), col] = random_sample.values

df
```
