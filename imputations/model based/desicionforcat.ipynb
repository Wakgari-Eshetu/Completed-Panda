```python
cat_cols = df.select_dtypes(include='object')
```

```python
cat_cols_encoded = pd.get_dummies(cat_cols.drop('Department' ,axis=1)).astype(int)
```

```python
num_cols = df.select_dtypes(include='number')
```

```python
df_cat_train  = pd.concat([num_cols , cat_cols_encoded] , axis= 1)
```

```python
cat_cols = pd.get_dummies(cat_cols).astype(int)
df = pd.concat([num_cols , cat_cols] , axis=1)
df
```

```python
df.isnull().sum()
```

```python
# imputation for numerical values 

missing_cols = ['Experience','Salary' ,'Rating']

for col in missing_cols:
    complete_data = df[df[col].notnull()]
    incomplete_data = df[df[col].isnull()]

    x_train = complete_data.drop(col , axis=1)
    y_train = complete_data[col]

    from sklearn.tree import DecisionTreeRegressor
    model = DecisionTreeRegressor()
    model.fit(x_train , y_train)

    predict = model.predict(incomplete_data.drop(col , axis=1))
    df.loc[df[col].isnull(), col] = predict

```

```python
df_check =df_cat_train.drop(['Experience' ,'Salary' ,'Rating'] , axis= 1)
df_check
columns_not_drop = ['Experience' ,'Salary' ,'Rating']
df = df.drop(columns= [c for c in df.columns if c not in columns_not_drop])
df

df = pd.concat([df  , df_check ] ,axis=1)
df
```

```python
df.isnull().sum()
```

```python
# for Categorical imputation of departement 
complete_cat = df[df['Department'].notnull()]
incomplete_cat = df[df['Department'].isnull()]

x_train = complete_cat.drop('Department' , axis=1)
y_train = complete_cat['Department']

from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train , y_train)

predict = model.predict(incomplete_cat.drop('Department' , axis= 1))

df.loc[df['Department'].isnull() , 'Department'] = predict

df
```

```python
df['Gender'] = df[['Gender_F' , 'Gender_M']].idxmax(axis=1)
df['Gender'] = df['Gender'].str.replace('Gender_' ,'')
df['Education'] = df[['Education_Bachelor','Education_Masters','Education_PhD']].idxmax(axis=1)
df['Education'] = df['Education'].str.replace('Education_','')
df['Position'] = df[['Position_Analyst', 'Position_Director', 'Position_Engineer',
       'Position_Executive', 'Position_Manager', 'Position_Senior Engineer',
       'Position_Senior Manager']].idxmax(axis=1)
df['Position'] = df['Position'].str.replace('Position_','')
df['Region'] = df[[ 'Region_East', 'Region_North',
       'Region_South', 'Region_West']].idxmax(axis=1)

df['Region'] = df['Region'].str.replace('Region_' ,'')

general_cols = ['Experience', 'Salary', 'Rating', 'Age', 'Tenure','Department', 'Gender', 'Education',
       'Position', 'Region']

df = df.drop(columns =[c for c in df.columns if c not in general_cols] , axis=1)
```
