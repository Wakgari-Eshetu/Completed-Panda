```python
import pandas as pd
import numpy as np

np.random.seed(42)

# Create a date index (100 days)
dates = pd.date_range(start="2024-01-01", periods=100, freq="D")

# Create synthetic data
data = {
    f"sensor{i}": np.random.normal(loc=50 + i*5, scale=10, size=100) 
    for i in range(1, 6)
}

# Add some irregular patterns
for i in range(1, 6):
    data[f"signal{chr(64+i)}"] = (
        np.linspace(10*i, 10*i + 100, 100) 
        + np.random.normal(scale=15, size=100)
    )

df = pd.DataFrame(data, index=dates)

# Introduce random missing values (~15-20%)
for col in df.columns:
    missing_indices = np.random.choice(df.index, size=int(0.17*len(df)), replace=False)
    df.loc[missing_indices, col] = np.nan

# Show first 10 rows
print(df.head(10))

```

```python
# using polynomial and spine etc 

df.interpolate(method='polynomial' , order= 3).bfill().ffill()
# or df.interpolate(method='spine',order = 2,3,4,5,7)
```
