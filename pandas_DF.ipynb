```python
# when creating from dict each key became the name of the column 
dicts = {"name":["leli", "kena","james","maother"] , 
         "age" :[5,6,7,8] , 
         "grade" :[11,22,33,44]}

df = pd.DataFrame( dicts, index=[x for x in range(1,len(dicts["name"])+1)])
print(df)
```

```python
# Pandas Dataframe indexing 
import pandas as pd

data = {'Name': ['John', 'Alice', 'Bob', 'Eve', 'Charlie'],
        'Age': [25, 30, 22, 35, 28],
        'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
        'Salary': [50000, 55000, 40000, 70000, 48000]}

df = pd.DataFrame(data, index= ['1','2','3','4','5'])
print(df.index)  # Accessing the index
```

```python
# i Can make the custom index from the key and values of dict 
# using set_index() function 
import pandas as pd

data = {'Name': ['John', 'Alice', 'Bob', 'Eve', 'Charlie'],
        'Age': [25, 30, 22, 35, 28],
        'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
        'Salary': [50000, 55000, 40000, 70000, 48000]}

df = pd.DataFrame(data, index= ['1','2','3','4','5'])
df = df.set_index('Name')
print(df)
```

```python
# If you need to reset the index back to default integer index,
#  use reset_index() method. This will convert the current 
# index into a regular column and create a new default index

df = df.reset_index()
print(df)
```

```python
# pandas access dataframe 
'''Accessing a dataframe in pandas involves retrieving, 
exploring, and manipulating data stored within this structure'''

# 1.  Accessing the column data frame 
df = [[1,2,3], [4,5,6], [7,8,9]]
df = pd.DataFrame(df)
print(df[0])  # is for default columns

df = [[1,2,3], [4,5,6], [7,8,9]]
df = pd.DataFrame(df , columns= ["one", "two", "three"])
print(df["one"]) # when the columns has specific name 
```

```python
# Accessing Rows by index using .loc[] and .iloc[] method 
# importing pandas as pd
import pandas as pd

# Creating the DataFrame
df = pd.DataFrame({'Weight': [45, 88, 56, 15, 71],
                   'Name': ['Sam', 'Andrea', 'Alex', 'Robin', 'Kia'],
                   'Age': [14, 25, 55, 8, 21]}, index= ['Row_1', 'Row_2', 'Row_3', 'Row_4', 'Row_5'])

# Print the DataFrame
print("Original DataFrame:")
print(df)

# Corrected selection using loc for a specific cell
result = df.loc['Row_2', 'Name'] # the output is Andrea

# Print the result
print("\nSelected Value at Row_2, Column 'Name':")
print(result)
```

```python
# we can select multiple rows and columns or rows or columns 
# importing pandas as pd
import pandas as pd

# Creating the DataFrame
df = pd.DataFrame({"A":[12, 4, 5, None, 1],
                "B":[7, 2, 54, 3, None],
                "C":[20, 16, 11, 3, 8],
                "D":[14, 3, None, 2, 6]})

# Create the index
index_ = ['Row_1', 'Row_2', 'Row_3', 'Row_4', 'Row_5']

# Set the index
df.index = index_

# Print the DataFrame
print("Original DataFrame:")
print(df)

# Corrected column names  in the result
result = df.loc[['Row_1','Row_2'],:] # every columns but the rows  row_1 and row_2 
result1 = df.loc['Row_1',:] # every columns but only the first row  
result2 = df.loc['Row_2':, :] # every columns and from row two to end both inclusive  
result3 = df.loc[:,:] # every both row and columns 
result4 = df.loc['Row_3':'Row_5',:] # every row and the columns is from b to end
result5 = df.loc[:'Row_2',:]  # every coulmns but the rows  are from the beg to Row_2 

# Corrected rows names  in the result
result = df.loc[:, ['A', 'D']] # every row but the columns from a and d 
result1 = df.loc[:,"A":"C"] # every row but the columns from a to c including c 
result2 = df.loc[:, "A"] # every row and one column called a 
result3 = df.loc[:,:] # every both row and columns 
result4 = df.loc[:,"B":] # every row and the columns is from b to end
result5 = df.loc[:,:"C"]  # every row but the columns are from the beg to c 
# Print the result
print("\nSelected Columns 'A' and 'D':")
print(result)

print("\nSelected Columns 'A' to 'C':")
print(result1)

print("\nSelected Column A:")
print(result2)

print("\nSelected Columns and row :")
print(result3)
```
